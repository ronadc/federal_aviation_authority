## **Data cleaning**

```python
required_columns = ["ACFT_MAKE_NAME","LOC_STATE_NAME","ACFT_MODEL_NAME","RMK_TEXT","FLT_PHASE", "EVENT_TYPE_DESC","FATAL_FLAG"]
```

```python
df = df_faa[required_columns]
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["FATAL_FLAG"].fillna(value="No",inplace=True)
```

```python
df.head()
```

```python
df_final = df.dropna(subset=["ACFT_MAKE_NAME"])
df_final.shape
```

```python
aircraftType = df_final.groupby("ACFT_MAKE_NAME")
```

```python
aircraftType.size()
```

```python
fatalAcc = df_final.groupby("FATAL_FLAG")
fatalAcc.size()
```

## **Display accidents where fatal flag is yes**

```python
acc_wit_fatality = fatalAcc.get_group("Yes")
acc_wit_fatality
```
